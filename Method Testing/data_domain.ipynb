#*In this notebook i am testing my custom methods for creating a domain space of a function using the tensorflow library*

```python
# Import libraries

import numpy as np
#%tensorflow_version 2.x
import tensorflow as tf
```

```python
# Create the initializer instances

interval_uniform_1d = tf.random_uniform_initializer(minval = -1, maxval = 1)

interval_uniform_2d = tf.random_uniform_initializer(minval = [-1,-2], maxval = [1, 2])

interval_uniform_3d = tf.random_uniform_initializer(minval = [-5, 1, 0], maxval = [6, 3, 1])

interval_normal = tf.random_normal_initializer(mean = 0.0, stddev = 1)
```

```python
interval_uniform_1d
```

```python
# Create samples with Tensors from the initializers with 100 sample points
  
sample_uniform_1d = interval_uniform_1d(shape = (100, 1), dtype = "float32")
sample_uniform_2d = interval_uniform_2d(shape = (100, 2), dtype = "float32")
sample_uniform_3d = interval_uniform_3d(shape = (100, 3), dtype = "float32")

sample_normal = interval_normal(shape = (100, 1), dtype = "float32")
```

```python
sample_uniform_1d
```

```python
vv = tf.Variable(sample_uniform_1d)
vv
```

```python
# Convert Tensors to numpy arrays for easier value access

array_uniform_1d = sample_uniform_1d.eval(session=tf.compat.v1.Session())
array_uniform_2d = sample_uniform_2d.eval(session=tf.compat.v1.Session())
array_uniform_3d = sample_uniform_3d.eval(session=tf.compat.v1.Session())

array_normal = sample_normal.eval(session=tf.compat.v1.Session())
```
